# 정규표현식과 토큰링

```python
from urllib.request import urlopen
import re # 정규표현식에 해당
```

```python
html = urlopen('http://www.naver.com')
print(html.read())
```

```python
from bs4 import BeautifulSoup # 복잡한 소스를 가독성 높이게 해준다.
```

```python
html = urlopen('http://www.naver.com')
bsObj = BeautifulSoup(html.read(), "html.parser")
```

```python
# 상당히 정리가 잘 되어 있는 형태로 만들 수 있다. 
print(bsObj.body.prettify())
```

```python
span_arr = bsObj.find_all('span') # <span>element를 포함하는 모든 태그들을 가져온다. 
span_arr
```

```python
link_arr = bsObj.find_all("a") # 링크할 떄 많이 사용된다.
link_arr
```

#### tag와 속성을 같이 탐색

```python
# tag안에 class이름이 blind라고 하는 tag만 찾아라 
span_blinds = bsObj.find_all("span", {"class" : 'blind'})
span_blinds
```

#### 형제객체 탐색

```python
# 옆에 있는 태그 검색 가능
for tmp in bsObj.body.find("input").next_siblings:
    print(tmp)
```

```python
# 옆에 있는 태그 검색 가능
for tmp in bsObj.body.find("input", {"id" : 'ie'}).next_siblings:
    print(tmp)
```

```python
# 이미지를 갖고 싶다면
bsObj.body.find_all("img")
```

#### 부모객체 탐색

```python
bsObj.body.find("img").parent
```

# 정규표현식

```python
# .jpg로 끝나는 사진만 찾아보기
bsObj.body.find_all("img", {"src" : re.compile(".jpg")})
```

```python
# https로 시작해서 png로 끝나는 사진만 찾아보기
bsObj.body.find_all("img", {"src" : re.compile("^(https).*png$")})
```

```python
# https로 시작해서 346을 포함하고 png로 끝나는 사진만 찾아보기
bsObj.body.find_all("img", {"src" : re.compile("^(https).*2020.*png$")})
```

#### 특수문자 제거

```python
text = "안녕하세요~!! ^^ 저는 토마스입니다..."
clean_text = re.sub("[!^@#$%&*():'';~.]", "", text) # 특수문자는 다 replace하라는 의미
clean_text
```

#### 정규식에 부합되는 부분 찾기

```python
re.search("[가-다]", "오늘의 날씨는 맑군요") #가에서 다 사이에 검색되는 단어 검색
```

```python
re.search("\d+", "안젤라는 1998년에 태어났습니다.") # 1개 이상의 숫자 의미
```
