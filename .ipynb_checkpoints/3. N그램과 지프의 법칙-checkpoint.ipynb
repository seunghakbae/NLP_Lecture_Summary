#### N그램 : 문장, 텍스트에서 N개의 연속된 요소들을 추출하는 방법

유니그램 : 한 개의 단어
바이그램 : 복합 단어
트리그램 : 세 개의 연속된 단어

#### 지프의 법칙
텍스트 뭉치에서 추출한 토큰 : 단어. 사용 빈도 높은 순서로 나열. ==> 순위 발생. 사용 빈도와 반비례. 

텍스트의 통계적 성질에 관한 법칙 중 하나로, 코퍼스 내에 등장하는 어휘의 빈도를 조사해서 크기순으로 정렬했을때, 그 빈도수와 순위는 반비례 관계에 있다는 법칙.



빈도 f는 순위r에 반비례한다. 



### 지프의 법칙 실습으로 알아보기

```python
from pylab import *
import nltk
nltk.download("stopwords")
from sklearn.datasets import fetch_20newsgroups
from nltk import word_tokenize
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
%matplotlib inline
import re
import string
from collections import Counter

newsgroups_data_sample = fetch_20newsgroups(subset='train')
```

```python
string.printable
```

```python
stop_words = stopwords.words('english')

stop_words
```

```python
tokenized_corpus = [word.lower() for sentence in newsgroups_data_sample['data'] for word in word_tokenize(re.sub(r'([^\s\w]|_)+', ' ', sentence)) if word.lower() not in stop_words]
```

```python
tokenized_corpus
```

```python
token_count_di = Counter(tokenized_corpus)
```

```python
token_count_di
```

```python
# 가장 많이 나오는 단어 50개
token_count_di.most_common(50)
```

```python
frequencies = [b for (a,b) in token_count_di.most_common(10000)]
tokens = [a for (a,b) in token_count_di.most_common(10000)]
ranks = range(1, len(frequencies) + 1)
```

```python
frequencies
```

```python
# 빈도 1위인 단어는 빈도 2위은 단어보다 약 2배 높은 빈도를 갖는다.
# 빈도 1위인 단어는 빈도 3위인 단어보다 약 3배 높은 빈도를 갖는다. 
```

```python
plt.figure(figsize=(8,8))
plt.ylim(1, 10**4)
plt.xlim(1, 10**4)

# Actual ranks and frequencies
obtained_line, = loglog(ranks, frequencies, marker = ".", label = 'Line obtained from the Text corpus')
obtained_legend = plt.legend(handles=[obtained_line], loc=1)
ax = plt.gca().add_artist(obtained_legend)
```

```python
# Expected ranks and frequencies as per Zipf's Law
expected_line, = plt.plot([1, frequencies[0]], [frequencies[0], 1], color = 'r', label = "line expected as per Zipf's Law")
plt.legend(handles=[expected_line], loc=4)

title("Plot starting Zipf law's in log-log scale")
xlabel("Rank of token in descending order of frequency of occurence")
ylabel("Frequency of occurrence of token")
grid(True)
```

# Extracting n-grams

```python
sentence = 'Sunil tweeted. "Witnessing 70th Republic Day of India from Rajpath. New Delhi. Mesmerizing performance by Indian Army! Awesome \
airshow! @india_official @indan_army #India #70thRepublic_Day For more photos ping me sunil@photoking.com :)"'
```

```python
import re

# n : gram의 개수
def n_gram_extractor(sentence, n):
    tokens = re.sub(r"([^\s\w]|_)+", " ", sentence).split()
    
    for i in range(len(tokens) - n + 1):
        print(tokens[i:i+n])
```

```python
# 바이그램
n_gram_extractor("The cute little boy is playing with kitten.", 2)
```

```python
# 트리그램
n_gram_extractor("The cute little boy is playing with kitten.", 3)
```

```python
# nltk에서 알아서 제공해주는 함수도 있다.
from nltk import ngrams
list(ngrams("The cute little boy is playing with kitten.".split(), 3))
```
